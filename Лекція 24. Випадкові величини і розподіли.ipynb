### –ó–∞–≤–¥–∞–Ω–Ω—è 1. *–í–∏–∑–Ω–∞—á–∏—Ç–∏ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ –π–æ–≥–æ —Å–∫–ª–∞–≤ –ø–µ—Ä—à–∏–π –µ–∫–æ–Ω–æ–º—ñ—Å—Ç*.
 
- [x] –î–≤–∞ –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∏ –∑–∞–ø–æ–≤–Ω—é—é—Ç—å –¥–æ–∫—É–º–µ–Ω—Ç–∏ –∑ –≤–ª–∞—Å–Ω–∏–º–∏ –ø—Ä–æ–≥–Ω–æ–∑–∞–º–∏ —â–æ–¥–æ –∫—É—Ä—Å—É –ø–µ–≤–Ω–∏—Ö –∫–æ—à—Ç–æ–≤–Ω–∏—Ö –ø–∞–ø–µ—Ä—ñ–≤, –ø—ñ—Å–ª—è —á–æ–≥–æ —ó—Ö —Å–∫–ª–∞–¥–∞—é—Ç—å —É —Å–ø—ñ–ª—å–Ω—É –ø–∞–ø–∫—É. –ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –ø—Ä–æ–≥–Ω–æ–∑–æ–≤–∞–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è –∫—É—Ä—Å—É –≤—ñ–¥ —Ñ–∞–∫—Ç–∏—á–Ω–æ–≥–æ –ø–æ–Ω–∞–¥ –¥–æ–ø—É—Å—Ç–∏–º—É –ø–æ—Ö–∏–±–∫—É –¥–ª—è 1-–≥–æ –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∞ –¥–æ—Ä—ñ–≤–Ω—é—î 0,1, –¥–ª—è 2-–≥–æ - 0,2. 1-–π –µ–∫–æ–Ω–æ–º—ñ—Å—Ç –∑–∞–ø–æ–≤–Ω–∏–≤ 40 –¥–æ–∫—É–º–µ–Ω—Ç—ñ–≤, 2-–π - 60 –¥–æ–∫—É–º–µ–Ω—Ç—ñ–≤. –ü—ñ–¥ —á–∞—Å –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ –Ω–∞–≤–º–∞–Ω–Ω—è –≤–∑—è—Ç–∏–π —ñ–∑ –ø–∞–ø–∫–∏ –¥–æ–∫—É–º–µ–Ω—Ç –≤–∏—è–≤–∏–≤—Å—è –∑ —Å—É—Ç—Ç—î–≤–æ—é –ø–æ—Ö–∏–±–∫–æ—é.
   * –í–∏–∑–Ω–∞—á–∏—Ç–∏ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ –π–æ–≥–æ —Å–∫–ª–∞–≤ –ø–µ—Ä—à–∏–π –µ–∫–æ–Ω–æ–º—ñ—Å—Ç.

$ P(Eco_1) $ - —Ü–µ –ø–æ–¥—ñ—è 1 (–¥–ª—è –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∞ 1). $ \\ $
$ P(Eco_2) $ - —Ü–µ –ø–æ–¥—ñ—è 2 (–¥–ª—è –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∞ 2). $ \\ $
$ –ê $ - –Ω–∞–º–≤–∞–Ω–Ω—è –≤–∑—è—Ç–∏–π —ñ–∑ –ø–∞–ø–∫–∏ –¥–æ–∫—É–º–µ–Ω—Ç.

*–í–∏–∑–Ω–∞—á–∞—é –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∞*: $ \\ $
$ P(Eco_1) = \frac{40}{100} = 0,4 $ $ \\ $
$ P(Eco_2) = \frac{60}{100} = 0,6 $

```python
eco_1 = 40 / 100
print('ùê∏ùëêùëú_1 =', eco_1)

eco_2 = 60 / 100
print('ùê∏ùëêùëú_2 =', eco_2)
```

$ P(A|Eco_1) = 0,1 $ (–∑–∞ —É–º–æ–≤–æ—é –∑–∞–¥–∞—á—ñ –ø–æ—Ö–∏–±–∫–∞ –¥–ª—è 1-–≥–æ –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∞ = 0,1) $ \\ $
$ P(A|Eco_2) = 0,2 $ (–∑–∞ —É–º–æ–≤–æ—é –∑–∞–¥–∞—á—ñ –ø–æ—Ö–∏–±–∫–∞ –¥–ª—è 2-–≥–æ –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∞ = 0,2)

–ó–∞ *—Ñ–æ—Ä–º—É–ª–æ—é –ø–æ–≤–Ω–æ—ó –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ* –≤–∏–∑–Ω–∞—á–µ–º–æ –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç: $ \\ $
$ P(A) = P(Eco_1)P(A|Eco_1) + P(Eco_2)P(A|Eco_2) = 0,4 ‚ãÖ 0,1 + 0,6 ‚ãÖ 0,2 = 0,16 $ $ \\ $
–ö–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç = $ 0,16 $

–ó–∞ *—Ñ–æ—Ä–º—É–ª–æ—é –ë–∞–π—î—Å–∞* –≤–∏—Ä–∞—Ö—É—î–º–æ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –∑–∞–ø–æ–≤–Ω–µ–Ω–Ω—è –¥–æ–∫—É–º–µ–Ω—Ç–∞ —ñ–∑ —Å—É—Ç—Ç—î–≤–æ—é –ø–æ—Ö–∏–±–∫–æ—é –ø–µ—Ä—à–∏–º –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–æ–º: $ \\ $
$ P(Eco_1|A) = \frac{P(Eco_1)P(A|Eco_1)}{P(A)} = \frac{0,4 ‚ãÖ 0,1}{0,16} = 0,25 $ $ \\ $

**–û—Ç–∂–µ, –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –¥–ª—è –ø–µ—Ä—à–æ–≥–æ –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∞ $ P(Eco_1|A) $ —Å—Ç–∞–Ω–æ–≤–∏—Ç—å** $ 0,25 $

```python
p_a = 0.4 * 0.1 + 0.6 * 0.2
print('–ö–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç =', p_a)

p_eco_1 = 0.4 * 0.1 / 0.16
print('–ô–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –¥–ª—è –ø–µ—Ä—à–æ–≥–æ –µ–∫–æ–Ω–æ–º—ñ—Å—Ç–∞ =', p_eco_1)
```

### –ó–∞–≤–¥–∞–Ω–Ω—è 2. *–û—Ü—ñ–Ω–∏—Ç–∏ —Å—Ç—É–ø—ñ–Ω—å —Ä–∏–∑–∏–∫—É —ñ –∑–∞–ø—Ä–æ–ø–æ–Ω—É–≤–∞—Ç–∏ —Ä—ñ—à–µ–Ω–Ω—è —â–æ–¥–æ –≤–∏–ø—É—Å–∫—É —ñ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó –æ–¥–Ω–æ–≥–æ –∑ –Ω–∞–±–æ—Ä—ñ–≤ —Ç–æ–≤–∞—Ä—ñ–≤*.

- [x] –Ñ –º–æ–∂–ª–∏–≤—ñ—Å—Ç—å –æ–±—Ä–∞—Ç–∏ —Å–ø–æ—Å—ñ–± –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ —ñ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó –¥–≤–æ—Ö –Ω–∞–±–æ—Ä—ñ–≤ —Ç–æ–≤–∞—Ä—ñ–≤ —à–∏—Ä–æ–∫–æ–≥–æ –≤–∂–∏—Ç–∫—É. –ó–∞ –¥–∞–Ω–∏–º–∏ –≤—ñ–¥–¥—ñ–ª—É –º–∞—Ä–∫–µ—Ç–∏–Ω–≥—É, —è–∫–∏–º –±—É–ª–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω—ñ –¥–æ—Å–ª—ñ–¥–∂–µ–Ω–Ω—è —Ä–∏–Ω–∫—É, –º–æ–∂–ª–∏–≤–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –≤—ñ–¥ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ —ñ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó X —Ç–∞ Y –Ω–∞–≤–µ–¥–µ–Ω–æ —É —Ç–∞–±–ª–∏—Ü—è—Ö (X, Y - –ø—Ä–∏–±—É—Ç–æ–∫ —É –≥—Ä–æ—à–æ–≤–∏—Ö –æ–¥–∏–Ω–∏—Ü—è—Ö) –Ω–∏–∂—á–µ.
    * –û—Ü—ñ–Ω–∏—Ç–∏ —Å—Ç—É–ø—ñ–Ω—å —Ä–∏–∑–∏–∫—É —ñ –∑–∞–ø—Ä–æ–ø–æ–Ω—É–≤–∞—Ç–∏ —Ä—ñ—à–µ–Ω–Ω—è —â–æ–¥–æ –≤–∏–ø—É—Å–∫—É —ñ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó –æ–¥–Ω–æ–≥–æ –∑ –Ω–∞–±–æ—Ä—ñ–≤ —Ç–æ–≤–∞—Ä—ñ–≤.
---

|*–¢–∞–±–ª–∏—Ü—è 1*|  |      |       |
|:---:  | :---:|:---: |:---:  |
| **X** | 1000 | 1500 |  2000 |
| **P** | 0,5  | 0,3  |  0,2  |

----
|*–¢–∞–±–ª–∏—Ü—è 2*|  |      |       |
|:---:  | :---:|:---: |:---:  |
| **Y** | 1000 | 1500 |  1750 |
| **P** | 0,4  | 0,4  |  0,2  |

----

    


$ X, Y $ - –≤–∏–ø–∞–¥–∫–æ–≤–∞ –≤–µ–ª–∏—á–∏–Ω–∞.

$ P $ - –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏.

C—É–º–∞ –≤—Å—ñ—Ö –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç–µ–π —É —Ç–∞–±–ª–∏—Ü—è—Ö 1 —Ç–∞ 2 –º–∞—î –¥–æ—Ä—ñ–≤–Ω—é–≤–∞—Ç–∏ $ P = 1 $, —Ç–æ–±—Ç–æ 1:

$ P(X) = 0,5 + 0,3 + 0,2 = 1 $ $ \\ $
$ P(Y) = 0,4 + 0,4 + 0,2 = 1  $

–û—Ç–æ–∂ —É–º–æ–≤–∞ –¥–ª—è –ø–æ–¥–∞–ª—å—à–æ–≥–æ —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É —î –ø—Ä–∞–≤–∏–ª—å–Ω–æ—é.

$ M(X) $ - –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ —Å–ø–æ–¥—ñ–≤–∞–Ω–Ω—è.

*–í–∏–∑–Ω–∞—á–∞—é –ú–ê–¢–ï–ú–ê–¢–ò–ß–ù–ï –°–ü–û–î–Ü–í–ê–ù–ù–Ø –º–æ–∂–ª–∏–≤–æ–≥–æ –ø—Ä–∏–±—É—Ç–∫—É –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –∑ –Ω–∞–±–æ—Ä—ñ–≤ —Ç–æ–≤–∞—Ä—ñ–≤*:

–í—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ —Ñ–æ—Ä–º—É–ª–∏ $ \sum^n _{i=1} X _i P_i $ :

$ M(X) = 1000 ‚ãÖ 0,5 + 1500 ‚ãÖ 0,3 + 2000 ‚ãÖ 0,2 = 1350 $ $ \\ $
–í—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –≤—ñ–¥ –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–æ–≥–æ —Å–ø–æ–¥—ñ–≤–∞–Ω–Ω—è –¥–ª—è $ M(X) = X - M(X) = 1000 - 1350 = -350; 1500 - 1350 = 150; 2000 - 1350 = 650. $

$ M(Y) = 1000 ‚ãÖ 0,4 + 1500 ‚ãÖ 0,4 + 1750 ‚ãÖ 0,2 = 1350 $ $ \\ $
–í—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –≤—ñ–¥ –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–æ–≥–æ —Å–ø–æ–¥—ñ–≤–∞–Ω–Ω—è –¥–ª—è $ M(Y) = Y - M(Y) = 1000 - 1350 = -350; 1500 - 1350 = 150; 1750 - 1350 = 400. $

#### –î–ª—è –∫–æ–∂–Ω–æ–≥–æ –Ω–∞–±–æ—Ä—É —Ç–æ–≤–∞—Ä—ñ–≤ –ø—Ä–∏–±—É—Ç–æ–∫ –≤–∏—è–≤–∏–≤—Å—è –æ–¥–Ω–∞–∫–æ–≤–∏–º: $ 1350 $

*–ó–Ω–∞—Ö–æ–¥–∂—É –î–ò–°–ü–ï–†–°–Ü–Æ –≤–∏–ø–∞–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ —Ç–∞ –æ—Ü—ñ–Ω—é—é —Å—Ç—É–ø—ñ–Ω—å —Ä–∏–∑–∏–∫—É* ***–ø–µ—Ä—à–æ–≥–æ –Ω–∞–±–æ—Ä—É***: 

$ D(X) = M(X^2) - (M(X)^2) = (1000 - 1350)^2 ‚ãÖ 0,5 + (1500 - 1350)^2 ‚ãÖ 0,3 + (2000 - 1350)^2 ‚ãÖ 0,2 = 
\\ = 61250 + 6750 + 84 500 = 152500 $ $ \\ $

##### –†–ò–ó–ò–ö –¥–ª—è –ø–µ—Ä—à–æ–≥–æ –Ω–∞–±–æ—Ä—É –∑ —Ç–∞–±–ª–∏—Ü—ñ 1: 
$ œÉ(X) = \sqrt{D(X)} = \sqrt{152500} = 390,5 $

*–ó–Ω–∞—Ö–æ–¥–∂—É –î–ò–°–ü–ï–†–°–Ü–Æ –≤–∏–ø–∞–∫–æ–≤–æ—ó –≤–µ–ª–∏—á–∏–Ω–∏ —Ç–∞ –æ—Ü—ñ–Ω—é—é —Å—Ç—É–ø—ñ–Ω—å —Ä–∏–∑–∏–∫—É* ***–¥—Ä—É–≥–æ–≥–æ –Ω–∞–±–æ—Ä—É***: 

$ D(Y) = M(Y^2) - (M(Y)^2) = (1000 - 1350)^2 ‚ãÖ 0,4 + (1500 - 1350)^2 ‚ãÖ 0,4 + (1750 - 1350)^2 ‚ãÖ 0,2 = 
\\ = 49000 + 9000 + 32000 = 90000 $ $ \\ $

##### –†–ò–ó–ò–ö –¥–ª—è –¥—Ä—É–≥–æ–æ –Ω–∞–±–æ—Ä—É –∑ —Ç–∞–±–ª–∏—Ü—ñ 2: 
$ œÉ(Y) = \sqrt{D(Y)} = \sqrt{90000} = 300 $

### –í–∏—Å–Ω–æ–≤–æ–∫

–ù–∞–±—ñ—Ä 1 –∑ —Ç–∞–±–ª–∏—Ü—ñ 1 –∑ —Ä–∏–∑–∏–∫–æ–º $ 390,5 $ –±—ñ–ª—å—à —Ä–∏–∑–∏–∫–æ–≤–∞–Ω–∏–π —É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤—ñ, –Ω—ñ–∂ –Ω–∞–±—ñ—Ä 2 –∑ —Ç–∞–±–ª–∏—Ü—ñ 2 –∑ —Ä–∏–∑–∏–∫–æ–º $ 300 $.

**–†–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –¥–ª—è –≤–∏—Ä–æ–±–Ω–∏—Ç–≤–∞ —ñ –ø—Ä–æ–¥–∞–∂—É —Ç—Ä–µ–±–∞ –Ω–∞–±—ñ—Ä 2.**
